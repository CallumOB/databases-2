## 2.	Check and tidy the data
 - Tidy sources:
     - From **imdf**, remove the columns 'Poster_Link',  'Certificate',  'Overview', 'No_of_Votes' and 'Gross'
     - From **actordf** remove the columns 'Place of Birth'
     - Copy **actordf** into a new dataframe, **adf**.

```python
im_discardcols = ['Poster_Link', 'Certificate', 'Overview', 'No_of_Votes', 'Gross']
imdf.drop(columns=im_discardcols, inplace=True)
imdf.head()
```

```python
actordf.drop(columns='Place of Birth', inplace=True)
actordf.head()
```

```python
adf = actordf
adf.head()
```

 - Rearrange the data:
     - Using **adf**, create a row for each performance as follows.  The 'Greatest Performances' column has a comma-separated list of stars. Create a row per actor, per performance as below:

     - Make sure none of the performance are empty - replace empty values with 'None'.
     - Add a new column, containing a list of the separated values.
     - Make a new dataframe, with a column for every list item.
     

```python
adf_split = adf['Greatest Performances'].str.split(",")
for i in adf_split:
    for j in i:
        j.replace("", "None")

adf['Performance List'] = adf_split
adf['Performance List'] = adf['Performance List'].astype(str)
adf.head()

adf_split[0][0]

new_df = adf
for i in adf_split:
    for j in range(len(i)):
        perf_index = j + 1
        new_df[f'Performance {perf_index}'] = i[j]

new_df.head()
```

 - Use a horizontal concatenation, to add the list columns to the original dataframe. Drop any redundant columns.
 - Melt the new columns down into a dataframe longadf, giving a row for each element of the list, calling the new column 'Performance'.
 - Drop the 'variable' column.
 - Pick one of the actor's names.  Check that there is a row in longadf for every performance that actor had in adf.
 - Only keep rows where 'Performance' does not have the value ''.

```python
longadf = pd.melt(
    adf, id_vars='Name',
    value_vars='Greatest Performances',
    value_name='Performance'
)
longadf.head()
perf_split = longadf.Performance.str.split(",", expand=True)
perf_split.replace("", "None")
perf_split.columns=['Performance 1', 'Performance 2', 'Performance 3', 'Performance 4', 'Performance 5', 'Performance 6']
adfl_parsed = pd.concat([longadf, perf_split], axis=1)
adfl_parsed.drop(columns=['variable', 'Performance'], inplace=True)
adfl_parsed.head()
```

- Now tackle imdf:
    - melt imdf to create a DataFrame **imdfl** with a row for each star from the 'Star1' to 'Star4' columns, calling the new column 'Starring'
    - drop the variable column from the resultant **imdfl**

```python
imdfl = pd.melt(
    imdf, id_vars=['Series_Title'],
    value_vars=['Star1', 'Star2', 'Star3', 'Star4'],
    value_name='Starring'
)
imdfl.drop(columns='variable', inplace=True)
imdfl.head()
```

## 3. Merge  the data

 - Merge the original actor dataframe **actordf** with **imdfl** on the actor's name, keeping all data.

```python
merge_imdf_actor = pd.merge(actordf, imdfl, left_on="Name", right_on="Starring")
```
